```python
import numpy as np
import re
from scipy import spatial
```

```python
#constants
size = 10000
```

```python
with open("clean.txt") as file:
    contents = file.read()

contents = re.sub(r'[^\w\s]',' ', contents)
contents = re.sub(' +', ' ', contents)
```

```python
characters = set(contents)
```

```python
table = {}
np.random.seed(76)
for c in characters:
    table[c] = np.random.choice([-1, 1], size=(size,))
```

```python
merge_order = []
```

```python
while len(table) > 2:
    context = {}
    for c in table:
        context[c] = np.zeros(size)
        
    del(context[' '])
    
    for i in range(len(contents)):
        c = contents[i]
        if c in context: 
            context[c] = np.add(context[c], np.multiply(np.roll(table[contents[i - 1]], -1), 
                                                        np.roll(table[contents[i + 1]], 1)))
    for c in context:
        context[c] = context[c] / np.linalg.norm(context[c])
    
    maxi = float('-inf')
    argmax = 0
    for c in context:
        for d in context:
            if c != d:
                temp = np.dot(context[c], context[d])
                if temp > maxi:
                    maxi = temp
                    argmax = (c, d)
    
    merge_order += [argmax]
    print(argmax)
    
    one, two = argmax
    del(table[two])
    contents = re.sub(two, one, contents)
```
